```python
from matplotlib import pyplot as plt
import numpy as np
```

# 1)     Load the necessary package for plotting using pyplot from matplotlib. Example - Days(x-axis) represents 8 days and Speed represents a carâ€™s speed. Plot a Basic line plot between days and car speed, put x axis label as days and y axis label as car speed and put title Car Speed Measurement.

```python
days = [1,2,3,4,5,6,7,8]
speed = [60,62,61,58,56,57,46,63]
plt.plot(days,speed,'r',label='line 1')
plt.title('Car Speed Measurement')
plt.xlabel('Days')
plt.ylabel('Car speed')
```

# 2)Now to above car data apply some string formats  like line style example green dotted line, marker shape like +, change markersize, markerface color etc.

```python
days = [1,2,3,4,5,6,7,8]
speed = [60,62,61,58,56,57,46,63]
plt.plot(days,speed,'g',marker=".",linestyle='dashed',label='line 1',markersize=20)
plt.title('Car Speed Measurement')
plt.xlabel('Days')
plt.ylabel('Car speed')
```

# 3) Plot Axes Labels, Chart title, Legend, Grid in Car minimum, Maximum and average speed in 8 days.

```python
days=[1,2,3,4,5,6,7,8]
max_speed=[80,91,92,88,77,79,76,75]
min_speed=[42,43,40,42,33,36,34,35]
avg_speed=[46,58,57,56,40,42,41,36]
plt.plot(days,max_speed,label='maximum speed',color='r',markersize=20,marker='+',linestyle='dashed')
plt.plot(days,min_speed,label='minimum speed',color='g',markersize=20,marker='*',linestyle='dashed')
plt.plot(days,avg_speed,label='average speed',color='b',markersize=20,marker='.',linestyle='dashed')
plt.legend()
plt.xlabel('days')
plt.ylabel('car speed')
plt.grid('b')
```

# 4) Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.

```python
x = np.arange(0,4*np.pi,0.1)   # start,stop,step
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,4*np.pi-1,0.1)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y,x,z)
plt.xlabel('x values from 0 to 4pi')
plt.ylabel('sin(x) and cos(x)')
plt.title('Plot of sin and cos from 0 to 4pi')
plt.legend(['sin(x)', 'cos(x)'])
plt.show()
```
