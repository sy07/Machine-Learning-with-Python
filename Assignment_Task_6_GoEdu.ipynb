# 1)     Load the necessary package for plotting using pyplot from matplotlib. Example - Days(x-axis) represents 8 days and Speed represents a carâ€™s speed. Plot a Basic line plot between days and car speed, put x axis label as days and y axis label as car speed and put title Car Speed Measurement.

```python
days = [1,2,3,4,5,6,7,8]
speed = [60,62,61,58,56,57,46,63]
plt.plot(days,speed,'r',label='line 1')
plt.title('Car Speed Measurement')
plt.xlabel('Days')
plt.ylabel('Car speed')
```

# 3) Plot Axes Labels, Chart title, Legend, Grid in Car minimum, Maximum and average speed in 8 days.

```python
days=[1,2,3,4,5,6,7,8]
max_speed=[80,91,92,88,77,79,76,75]
min_speed=[42,43,40,42,33,36,34,35]
avg_speed=[46,58,57,56,40,42,41,36]
plt.plot(days,max_speed,label='maximum speed',color='r',markersize=20,marker='+',linestyle='dashed')
plt.plot(days,min_speed,label='minimum speed',color='g',markersize=20,marker='*',linestyle='dashed')
plt.plot(days,avg_speed,label='average speed',color='b',markersize=20,marker='.',linestyle='dashed')
plt.legend()
plt.xlabel('days')
plt.ylabel('car speed')
plt.grid('b')
```

# 4) Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.

```python
x = np.arange(0,4*np.pi,0.1)   # start,stop,step
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,4*np.pi-1,0.1)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y,x,z)
plt.xlabel('x values from 0 to 4pi')
plt.ylabel('sin(x) and cos(x)')
plt.title('Plot of sin and cos from 0 to 4pi')
plt.legend(['sin(x)', 'cos(x)'])
plt.show()
```

#5. Plot Simple bar chart showing popularity of Programming Languages.

Plot Multiple Bars showing Popularity and Security of major Programming Languages. Also Create Horizontal bar chart using barh function.

```python
import numpy as np
from matplotlib import pyplot as plt

languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
popularity = [56, 39, 34, 34, 29]
security = [44 ,36 ,55, 50, 42]

# create an array of len(langauges) size
x_axis = np.arange(len(languages))

#suppose x_axis = [0,1,2,3,4] then for the 1st bar graph first bar graph will draw at -0.2 (0-0.2) and second at 0.8(1-0.2)
#for secomd bar graph first bar will draw at 0.2 2nd at 1.2 , 3rd at 3.2 and so on
plt.bar(x_axis-0.2, popularity , label="Popularity" , color="r" ,width=0.4)
plt.bar(x_axis+0.2, security , label="Security" , color="b", width=0.4)

plt.xticks(x_axis, languages)
plt.xlabel("Languages")
plt.ylabel("Number")
plt.title("Bar graph with Multiple bar ")
plt.legend()
plt.show()
```

```python
import numpy as np

languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
popularity = [56, 39, 34, 34, 29]
security = [44 ,36 ,55, 50, 42]

# create an array of len(langauges) size
x_axis = np.arange(len(languages))

#suppose x_axis = [0,1,2,3,4] then for the 1st bar graph first bar graph will draw at -0.2 (0-0.2) and second at 0.8(1-0.2)
#for secomd bar graph first bar will draw at 0.2 2nd at 1.2 , 3rd at 3.2 and so on
plt.barh(x_axis-0.2, popularity , label="Popularity" , color="r" ,height=0.4)
plt.barh(x_axis+0.2, security , label="Security" , color="b", height=0.4)

plt.yticks(x_axis, languages)
plt.ylabel("Languages")
plt.xlabel("Number")
plt.title("Bar graph with Multiple bar")
plt.legend()
plt.show()
```

#7.  Titanic Data Set Download Data

 Load the data file

 (i) Create a pie chart presenting the male/female proportion 

 (ii) Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
import pandas as pd
titanic = pd.read_csv("/content/titanic_original.csv")
titanic
```

```python
titanic[0:10]
```

```python
titanic['sex']
```

```python
titanic.columns
```

```python
plt.figure(figsize=(10,10))
titanic.sex.value_counts().plot.pie(autopct='%.0f%%',labels=['Male','Female'],shadow = True,startangle = 180)
plt.title('Gender Proportion')
plt.show()
```

```python
male = titanic[titanic['sex'] == 'male']
male.sex.value_counts()
```

```python
female = titanic[titanic['sex'] == 'female']
female.sex.value_counts()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(male.fare,male.age,label='Male',color='red',marker='.')
plt.scatter(female.fare,female.age,label='Female',color='green',marker='+')
plt.xlabel('Fare')
plt.ylabel('Age')
plt.title("Plot Scattering by Fare and Age")
plt.legend()
plt.show()
```
